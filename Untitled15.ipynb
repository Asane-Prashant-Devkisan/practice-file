```python
#replacing missing values for children value with 0 and country with near 
# drop out the value of agent and companny 
replacing={'children':0,'country':'Mean'}
df_new=df.fillna(replacing)
df_new.head()
```

```python
#ropping
df_new.drop(['company','agent'],axis=1,inplace=True)
df_new.head
```

```python
df_new.isna().sum()
```

```python
#convert the datatype of childewn which from float to integer 
df_new['children'] = df_new['children'].astype(int)
df_new.info()
```

```python
#total gueust
df_new["total_guest"]=df_new["children"]+df_new["babies"]+df_new["adults"]

df_new.head()
```

```python
child_proportion=df_new["children"].sum()/df_new["total_guest"].sum()
child_proportion
```

```python
# Filter: with children vs without children
with_children = df[df['total_kids'] > 0]
without_children = df[df['total_kids'] == 0]

# Group by month and count bookings
with_kids_peak = with_children['arrival_date_month'].value_counts().sort_values(ascending=False)
without_kids_peak = without_children['arrival_date_month'].value_counts().sort_values(ascending=False)

# Display results
print("ðŸ“Œ Peak months with children:\n", with_kids_peak.head(3))
print("\nðŸ“Œ Peak months without children:\n", without_kids_peak.head(3))
```
