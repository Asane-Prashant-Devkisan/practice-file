# Visualization using matlotlib 


```python
#1 Create a line chart 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
x=np.arange(0,10,0.1)
y=2*x+5
plt.plot(x,y)
plt.show()
```

```python
#2 Customize the above line plot add title axix labels
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
x=np.arange(0,10,1)
y=2*x+5
plt.plot(x,y,linewidth=2.0,linestyle=":",color="r",alpha=0.5,marker='o')
plt.show()
```

```python
# 3 Plot 2 lines a single plot 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
x=np.arange(0,10,1)
y1=2*x+5
y2=3*x+10
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# 3 Plot 2 lines a single plot 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
x=np.arange(0,10,1)
y1=2*x+5
y2=3*x+10
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title("graph1")
plt.subplot(1,2,1)
plt.plot(x,y2)
plt.title("graph2")
plt.show()
```

```python
# 7Create a scatter plot 
import matplotlib.pyplot as plt 
%matplotlib inline 
a=[10,20,30,40,50,60,70,80,90]
b=[8,9,6,7,4,5,8,9,6]
x=[1,2,3,1,1,1,7,8,9]
#creating the scatter plot 
plt.scatter(a,b)
plt.scatter(a,x)
plt.show()
```

```python
# 8 Customizing the above scatter plot  
import matplotlib.pyplot as plt 
%matplotlib inline 
a=[10,20,30,40,50,60,70,80,90]
b=[8,9,6,7,4,5,8,9,6]
x=[1,2,3,1,1,1,7,8,9]
#creating the scatter plot 
plt.scatter(a,b,c='g',s=200,edgecolor='y',marker='o',alpha=0.5)
plt.scatter(a,x,c='r',s=400,edgecolor='b',marker='4',alpha=1)
plt.legend(['b','x'],loc='best')
plt.title("scatter plot demo")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# 9 how to create a histogram 
import matplotlib.pyplot as plt
%matplotlib inline 
numbers=[10,12,16,19,11,20,28,38,35,34,60,45,55,68,62,64,70,78,75,79,85,89,95]
plt.hist(numbers,bins=[0,20,40,55,60,80],edgecolor='b')
plt.show()
```

```python
#10 How to create a box plot 
import matplotlib.pyplot as plt 
%matplotlib inline 
total=[20,40,50,60,80,90]
order=[10,3,1,15,17,2,30,44,2,1]
discount=[20,10,20,5,10,20,50,60,20,45]
data=list([total,order,discount])
plt.boxplot(data,showmeans=True)
plt.title("Box plot demo")
plt.xlabel("Values")
plt.grid(True)
plt.show()
```

```python
# 11 Create a violin plot 
import matplotlib.pyplot as plt 
%matplotlib inline 
#plottting the data 
plt.violinplot(data,showmeans=True,showmedians=True)
plt.title("violin plot demo")
plt.grid(True)
plt.show()
```

```python
# 12 Create a pie chart 
#preparing the data 
label=['dog','cat','cow','buffallow']
sizes=[50,45,80,90]
plt.pie(sizes,labels=label)
plt.title("pie chart demo")
plt.show()
```

```python
# 13 Create a doughnut chart 
group_names=["GroupA","GroupB","GroupC"]
group_size=[2,3,4]
size_centre=[3]
a,b,c=[plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
pie1=plt.pie(group_size,labels=group_names,radius=1.0,colors=[a(0.5),b(1.5),c(1.0)])
pie2=plt.pie(size_centre,radius=0.5,colors='w')
plt.show()
```

```python
# 15  iMPORTING TITANIC DATASETS FROM  SEABORN LIBRARY
import seaborn as sns
import matplotlib.pyplot as plt 
data=sns.load_dataset("titanic")
data
```

```python
data['survived'].value_counts()
```

```python
# 16 Create a bar chart to show survived variable using matplotlib 
data['survived'].value_counts().plot(kind='bar')
plt.show()
```

```python
# 17 Create a bar chart to show survived variable using matplotlib 
data['survived'].value_counts().plot(kind='pie')
plt.show()
```

```python
# 18 Creating a pie chart for checking proportion of passengers in different classes 
data['pclass'].value_counts().plot(kind='pie')
plt.show()
```

```python
# 19 Creating a bar chart for checking proportion of passengers in different classes 
data['pclass'].value_counts().plot(kind='bar')
plt.show()
```

```python
#20 creating a pie chart & bar chart customized 
#firstly we will check how much passengers in sitted in each class 
pclass_counts=data['pclass'].value_counts().sort_index()
pclass_counts
```

```python
 # 21 PLotting bar chart 
pclass_counts=data['pclass'].value_counts().sort_index()
plt.figure(figsize=(6,4))
plt.bar(x=pclass_counts.index.astype(str),height=pclass_counts.values,color='skyblue',edgecolor='b')
plt.xlabel('passenger class')
plt.ylabel('number of passengers')
plt.title('passenger count by class')
plt.show()
```

```python
 # 22 Create a  Histogram 
# Histogram age distribution 
plt.figure(figsize=(8,5))
plt.hist(data['age'].dropna(),bins=10,color='lightblue',edgecolor='black')
plt.xlabel('passsenger class')
plt.ylabel('number of passengers')
plt.title('distribution of passenger age')
plt.show()
```

```python
# 23 Creating a pie chart in detail 
from typing_extensions import dataclass_transform 
#pie chart Gender _Distribution 
gender_counts=data['sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',colors=['pink','lightblue'])
plt.show()
```

```python
# 24  Creating a stack  chart in detail 
data.groupby(['pclass','survived']).size().unstack(fill_value=0)
```

```python
#25 Create a stacked bar chart 
survived_counts=data.groupby(['pclass','survived']).size().unstack(fill_value=0)
plt.figure(figsize=(8,5))
plt.bar(x=survived_counts.index.astype(str),height=survived_counts[1],label='survived',color='limegreen')
# Plot survived
#plt.bar(x=survived_counts.index.astype(str),height=survived_counts[1],label='Survived',color='limegreen')
plt.bar(survived_counts.index.astype(str),survived_counts[0],bottom=survived_counts[1],label='did not survive',color='salmon')
plt.xlabel('passenger class')
plt.ylabel('number of passengers')
plt.title('survival by class')
plt.legend()
plt.show()
```

```python
# 26 scatter plot Age vs Fare 
plt.figure(figsize=(8,6))
plt.scatter(data['age'],data['fare'],alpha=1,c='purple',edgecolor='w')
plt.xlabel('age')
plt.ylabel('fare')
plt.title('age vs fare')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get sorted list of unique passenger classes
pclasses = sorted(data['pclass'].dropna().unique())

# Prepare data for boxplot: list of fare values for each pclass
fare_data = [data[data['pclass'] == pc]['fare'].dropna() for pc in pclasses]

# Plot
plt.figure(figsize=(8, 5))
plt.boxplot(fare_data, labels=pclasses)
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.show()
```

```python
# Get sorted list of unique passenger classes
pclasses = sorted(data['pclass'].dropna().unique())

# Group fare values by pclass
data_by_class = [data[data['pclass'] == pc]['fare'].dropna() for pc in pclasses]

# Create the box plot
plt.figure(figsize=(8, 5))
plt.boxplot(data_by_class, labels=pclasses)

plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Fare Distribution by Passenger Class')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
